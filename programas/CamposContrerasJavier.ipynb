```python
def ejercicio1():
    t = 0; p = 0; contador = 0; contador2 = 0 # Variables importantes

    while True:
        try:
            num = int(input("Números de correos: "))
            if num > 0:
                break
        except ValueError:
            pass

    correos = []
    for x in range(num):
        correo = input(f"Ingrese el {x+1} correo: ")
        while True:
            if "@" in correo and "." in correo:
                m = correo.split(".")
                correo = m[-1]
                correos.append(correo)
                break
            else:
                correo = input(f"Ingrese el {x+1} correo: ")

    correos.sort() # Variables importantes

    for x in correos:
        if x == correos[t]:
            contador += 1
            k = x
        else:
            contador2 += 1
            if contador <= contador2:
                contador = contador2
                contador2 = 0
                t = p
            elif x == correos[p-1]:
                pass
            else:
                contador2 = 1
        p += 1
    print(f"{k} - {contador}")

ejercicio1() # Ejecutar de esa manera!.
```

```python
import random; import numpy as np

def ejercicio3():
    while True:
        try:
            columnas = int(input("Ingrese las dimensiones: "))
            if columnas > 0:
                break
        except ValueError:
            pass

    while True:
        try:
            filas = int(input("Ingrese las dimensiones: "))
            if filas > 0:
                break
        except ValueError:
            pass
    
    n = filas*columnas
    matriz1 = np.zeros((columnas, filas)) # Matriz principal
    peru = input("Quiere que los numeros lo coloque usted o la maquina (Escriba MAQUINA o YO): ").upper()
    match peru:
        case "MAQUINA":
            for x in range(n):
                while True:
                    eu = random.randrange(-9, 9) # ESTE COMANDO ES PARA USARLO CON UN RANDOMIZADOR
                    jaaj = True
                    while jaaj:
                        p = random.randrange(0, columnas)
                        k = random.randrange(0, filas)
                        if matriz1[p, k] == 0:
                            matriz1[p, k] = eu
                            jaaj = False
                        else:
                            pass
                    break
        case "YO":
            for x in range(columnas):
                for i in range(filas):
                    eu = int(input(f"Ingrese el {x+1} numero: ")) # random.randrange(-9, 9) ESTE COMANDO ES PARA USARLO CON UN RANDOMIZADOR
                    matriz1[x, i] = eu
    

    print(matriz1, "\n")


    matriz2 = []; index = 0 
    for x in range(columnas):
        for i in range(filas):
            n = matriz1[x, i]
            if n != 0:
                matriz2.append([])
                matriz2[index] = [x, i, n]
                index += 1
   
    print(np.array(matriz2))

ejercicio3() # Ejecutar de esa manera!. COMENTARLE QUE HICE EL CASO DE MANERA MANUAL Y PSEUDO-ALEATORIA POR NO ESPECIFICACION DEL PROBLEMA.
```
